```python
df[df['Receipt Number'] == 'GIJ00003'].Amount.sum()
```

Бачимо, що в таблиці сума по рахунку по показнику "Net Bill Amount" становить 38.3526, у той час як наш перерахунок дав результат 30.7692. Перерахуємо показники "Net Bill Amount" та "Gross Bill Amount" окремо для кожної позиції та додамо нові стовбчики до набору даних.

```python
# Calculate the sum of 'Value' for each 'Category'
amount_correct = df.groupby('Receipt Number')['Amount'].sum().reset_index()

# Rename the column to indicate it's a sum
amount_correct.rename(columns={'Value': 'Net Bill Amount Correct'}, inplace=True)

# Merge the sum back to the original DataFrame
df = df.merge(amount_correct, on='Receipt Number')
```

```python
df[:5]
```

```python
df = df.rename(columns={"Amount_x": "Amount"})
```

```python
df = df.rename(columns={"Amount_y": "Net Bill Amount Correct"})
```

```python
df['Gross Bill Amount Correct'] = df['Net Bill Amount Correct'] * (1 + df.GST)
```

```python
df[:5]
```
